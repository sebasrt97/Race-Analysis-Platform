```python
import pandas as pd

# 1. Cargar el JSON generado por Scrapy
df = pd.read_json('/home/bigdata/Race-Analysis-Platform/scrapy_project/ediciones.json')


# 3. Extraer el año de la fecha (asumiendo formato DD/MM/YYYY)
df['year'] = pd.to_datetime(df['fecha'], format='%d/%m/%Y').dt.year
```

```python
df['finish_time_normalizado'] = df['finish_time'] - df['finish_time'].dt.normalize()

df['finish_time_segundos'] = df['finish_time_normalizado'].dt.total_seconds()

df['finish_time_segundos'] = pd.to_timedelta(df['finish_time_segundos'], unit='s').dt.floor('s').astype(str).str.split(' ').str[-1]
```

## Average finish times by age group 

```python
print(df.groupby('gender')['finish_time_segundos'].mean())

```

## Average finish times by gender 

```python
print(df.groupby('gender')['finish_time_segundos'].mean())
```

```python
tendencia_tiempos = df.groupby('year')['finish_time_seconds'].agg(['min', 'max']).reset_index()

tendencia_tiempos['min_hms'] = pd.to_timedelta(tendencia_tiempos['min'], unit='s').dt.floor('s').astype(str).str.split(' ').str[-1]
tendencia_tiempos['max_hms'] = pd.to_timedelta(tendencia_tiempos['max'], unit='s').dt.floor('s').astype(str).str.split(' ').str[-1]
tendencia_tiempos.head()
```
